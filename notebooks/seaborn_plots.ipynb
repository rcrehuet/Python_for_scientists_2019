```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the example tips dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.violinplot(data=tips, x='day', y='tip',palette='Set2')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex',
          palette='inferno')
```

```python
tips['tip_ratio'] = tips['tip']/tips['total_bill']
tips.head()
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip_ratio')
```

```python
import numpy as np
```

```python
points = np.random.normal(loc=5, size=100)
```

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
sns.boxplot(x="day", y="total_bill", hue="weekend", data=tips, dodge=True);
```

```python
sns.kdeplot(points, shade=True,)
sns.kdeplot(points-3, shade=True)
plt.xlabel("temp.")
```
